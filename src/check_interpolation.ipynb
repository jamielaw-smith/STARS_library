```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os
exec(open("/Users/lawsmith/Dropbox/jYT/my_settings.py").read())
```

```python
stardirs = os.listdir('../output/')
stardirs.sort()
for stardir in stardirs:
    #print(stardir)
    fig, ax = plt.subplots()

    betafiles = os.listdir('../output/'+stardir+'/')
    betafiles.sort()
    for betafile in betafiles:
        #print(betafile)
        t, dmdt = np.loadtxt('../output/'+stardir+'/'+betafile, unpack=True)
        ax.plot(t, dmdt, c='r', lw=2)

    betafiles = os.listdir('../input/'+stardir+'/')
    betafiles.sort()
    for betafile in betafiles:
        t, dmdt = np.loadtxt('../input/'+stardir+'/'+betafile, unpack=True, skiprows=1)
        ax.plot(t, dmdt, c='k', lw=2)
    
    ax.set_title(stardir)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.grid()
```

```python
stardirs = os.listdir('../output/')
stardirs.sort()
fig,ax = plt.subplots()
fig2,ax2 = plt.subplots()
for stardir in stardirs:
    if stardir == '.DS_Store':
        continue
    betafiles = os.listdir('../output/'+stardir+'/')
    betafiles.sort()
    for betafile in betafiles:
        if betafile == '.DS_Store':
            continue
        t, dmdt = np.loadtxt('../output/'+stardir+'/'+betafile, unpack=True)
        ax.plot(t, dmdt, c='k')
        ax2.plot(t/t[np.argmax(dmdt)], dmdt/max(dmdt), c='k')

    betafiles = os.listdir('../input/'+stardir+'/')
    betafiles.sort()
    for betafile in betafiles:
        if betafile == '.DS_Store':
            continue
        t, dmdt = np.loadtxt('../input/'+stardir+'/'+betafile, unpack=True, skiprows=1)
        ax.plot(t, dmdt, c='r')
        ax2.plot(t/t[np.argmax(dmdt)], dmdt/max(dmdt), c='r')
    
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid()
ax.set_xlim(3, 1e3)
ax.set_ylim(1e-4, 1e1)

ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.grid()
ax2.set_xlim(0.3, 10)
ax2.set_ylim(1e-2, 1.05)
```
